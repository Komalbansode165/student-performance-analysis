Student Performance Analysis using Python


Introduction

This project focuses on analyzing the academic performance of 25 students using Python libraries such as Pandas, NumPy, and Matplotlib. The dataset contains marks for four subjects: Math, Science, English, and Attendance. 

The objective of this analysis is to understand the overall performance of the class by calculating summary statistics such as the average, maximum, and minimum marks for each subject. 
To present the data clearly, the grade order was organised properly, and an additional column “No” was added to reset the indexing from 1 for easier readability. The project also identifies the topper using the query() function and visualizes key findings through a bar chart of average subject marks and a pie chart showing grade distribution.

This analysis provides meaningful insights into student performance trends and helps understand strengths, weaknesses, and overall class outcomes.



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("students25.csv")
df
```

Calculated Average of the subjects

```python
maths_average=df["Math"].mean()
maths_average
```

```python
english_average=df["English"].mean()
english_average
```

```python
science_average=df["Science"].mean()
science_average
```

Calculated highest & lowest marks

```python
highest_math = df["Math"].max()
lowest_math = df["Math"].min()

highest_math, lowest_math
```

```python
highest_science = df["Science"].max()
lowest_science = df["Science"].min()

highest_science, lowest_science
```

```python
highest_english = df["English"].max()
lowest_english = df["English"].min()

highest_english, lowest_english
```

fix the grade order & add a column No to start indexing from 1

```python
grade_order=["A+","A","B+","B","C","D"]
df["Grade"] = pd.Categorical(df["Grade"], categories=grade_order, ordered=True)

sorted_df=df.sort_values(by=["Grade","Percentage"],ascending=[True,False])

sorted_df = sorted_df.reset_index(drop=True)         
sorted_df.insert(0, "No", range(1, len(sorted_df) + 1)) 
sorted_df

```

find the topper using query()

```python
topper=df.query("Percentage > 90 and Grade =='A+'")
topper
```

```python
df.describe()
```

Bar Chart of Average Subject Marks

```python
subject_avg = df[['Math','Science','English','Attendance']].mean()

plt.figure(figsize=(7,5))
plt.bar(subject_avg.index, subject_avg.values)
plt.title("Average Marks of Subjects")
plt.ylabel("Marks")
plt.show()

```

Pie Chart of Grade Distribution

```python
grade_counts = df["Grade"].value_counts()
plt.figure(figsize=(6,6))
explode=[0,0,0,0,0.1]
plt.pie(grade_counts, labels=grade_counts.index,startangle=140, autopct='%1.1f%%',explode=explode,shadow=True)
plt.title("Grade Distribution")
plt.show()

```
